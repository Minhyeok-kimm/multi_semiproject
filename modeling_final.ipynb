# Modeling

## 침수 수위 예측(Regressor)

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# 평가지표 함수 생성
def adjusted_r_squared(r2_score,X_test):
    	return 1 - (1-r2_score)*(len(X_test)-1)/(len(X_test)-X_test.shape[1]-1)

def regressor_metrics(y_test, y_pred):
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    print(f'MAE: {mae:.4f}')
    print(f'MSE: {mse:.4f}')
    print(f'RMSE: {rmse:.4f}')
    print(f'R2: {r2:.4f}')
```

```python
# x, y 분리 함수 생성
def split_xy(df, encoding, target):
    from sklearn.model_selection import train_test_split
    if encoding == 'onehot':
        if target == 'avg':
            df_x = df.iloc[:, [i for i in range(len(df.columns)) if i not in (11, 12)]]
            df_y = df.iloc[:, 11]
            return df_x, df_y
        elif target == 'max':
            df_x = df.iloc[:, [i for i in range(len(df.columns)) if i not in (11, 12)]]
            df_y = df.iloc[:, 12]
            return df_x, df_y
        else:
            print('target is avg or max')
    elif encoding == 'label':
        if target == 'avg':
            df_x = df.iloc[:, [i for i in range(len(df.columns)) if i not in (12, 13)]]
            df_y = df.iloc[:, 12]
            return df_x, df_y
        elif target == 'max':
            df_x = df.iloc[:, [i for i in range(len(df.columns)) if i not in (12, 13)]]
            df_y = df.iloc[:, 13]
            return df_x, df_y
        else:
            print('target is avg or max')
    else:
        print('encoding is onehot or label')
```

### 모델 학습 수행

```python
# 모델학습 함수 생성
# 주 타깃인 onehot - max를 기준으로 생성
def model_train(model):
    df_x, df_y = split_xy(df_floor_onehot, 'onehot', 'max')
    
    del df_x['경사도']
    del df_x['빗물받이개수']
    del df_x['일평균강수량']
    
    train_x, test_x, train_y, test_y = train_test_split(df_x, df_y, test_size=0.2, random_state=0, shuffle=True)

    model = model
    model.fit(train_x, train_y)
    predict_y = model.predict(test_x)
    regressor_metrics(test_y, predict_y)
```

```python
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor
print('RandomForestRegressor(random_state=0)')
model_train(RandomForestRegressor(random_state=0))
print('LGBMRegressor(random_state=0, force_col_wise=True)')
model_train(LGBMRegressor(random_state=0, force_col_wise=True))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
print('LinearRegression()')
model_train(LinearRegression())
print('Lasso()')
model_train(Lasso())
```

```python
print('Lasso(alpha=0.0001)')
model_train(Lasso(alpha=0.0001))
print('Lasso(alpha=10)')
model_train(Lasso(alpha=10))
```

```python
from sklearn.linear_model import ElasticNet
print('ElasticNet')
model_train(ElasticNet())
```

```python
from sklearn.model_selection import GridSearchCV

param_lgb = {
    "learning_rate" : [0.01, 0.1, 0.2, 0.3, 0.5],
    # "max_depth" : [25, 50, 75],
    "num_leaves" : [300,500,900,1200],
    "n_estimators" : [200, 300, 500, 800],
}

df_x, df_y = split_xy(df_floor_onehot, 'onehot', 'max')

del df_x['경사도']
del df_x['빗물받이개수']
del df_x['일평균강수량']

train_x, test_x, train_y, test_y = train_test_split(df_x, df_y, test_size=0.2, random_state=0, shuffle=True)

gs_lb = GridSearchCV(LGBMRegressor(random_state=0, n_jobs=-1, force_col_wise=True, verbose=-1), param_grid=param_lgb, cv=3)
gs_lb.fit(train_x, train_y)
```

```python
def gridsearchcv_model(model, params, cvs=3):
    from sklearn.model_selection import GridSearchCV
    df_x, df_y = split_xy(df_floor_onehot, 'onehot', 'max')
    
    del df_x['경사도']
    del df_x['빗물받이개수']
    del df_x['일평균강수량']
    
    train_x, test_x, train_y, test_y = train_test_split(df_x, df_y, test_size=0.2, random_state=0, shuffle=True)
    
    gs = GridSearchCV(model, param_grid=params, cv=cvs)
    gs.fit(train_x, train_y)
    
    print(f'best_params: {gs.best_params_}')
    
    pred = gs.predict(test_x)
    regressor_metrics(test_y, pred)
    
    return gs
```

```python
from sklearn.ensemble import RandomForestRegressor
param_rfr = {
    "max_depth" : [25, 50, 75], # 과적합 방지
    "n_estimators" : [200, 300, 500, 800],
    "min_samples_split":[2, 3, 5, 8],
    "min_samples_leaf":[1, 2, 3, 5, 8]
}

gs_rf = gridsearchcv_model(RandomForestRegressor(random_state=0, n_jobs=-1), param_rfr)
```
